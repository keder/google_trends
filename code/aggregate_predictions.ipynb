```python
import pandas as pd
import os
import os.path
import re
from scipy.special import inv_boxcox
```

```python
aggregate_path = "../Temp"
data_frames = []
pattern = re.compile(r"ARIMA\-")

for root, dirs, files in os.walk(aggregate_path):
    for file in files:
        if pattern.match(file):
            file_path = os.path.join(root, file)
            df = pd.read_csv(file_path)
            data_frames.append(df)

# Concatenate all data frames into a single data frame
combined_df = pd.concat(data_frames, ignore_index=True)
```

```python
combined_df
```

```python
# combined_df = combined_df.apply(lambda col: inv_boxcox(col, 0.5))
combined_df["mean_trans"] = inv_boxcox(combined_df["means"], 0.5)
```

```python
combined_df.to_csv(os.path.join(aggregate_path, "dashboard_prediction.csv"), index=False)
```
